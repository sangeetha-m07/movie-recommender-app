### step 2 : data cleaning and feature selection

```python
movies = movies.merge(credits, on = "title")           ## here we merged the movies and credits into single dataset
```

```python
movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]   ## keeping the required columns 
```

```python
## genres : Genres mean the category or type of the movie based on its theme, style, or story
## keywords : Keywords are important words or tags that describe the main themes, topics, or concepts of a movie
## overview : Overview is a short description or summary of what the movie is about â€” like the plot
## crew : People who work behind the scenes to make the movie, but donâ€™t act in it(director, producer, writer , composer etc...)
```

```python
movies.rename(columns = {"id" : "movie_id"}, inplace = True)    ## renaming the id with movie_id(if id is present instead of movie_id)
```

### step 3 : Data transformation  && step 4 : create tags for each movie

```python
import ast

# Safer functions to handle bad data
def safe_convert(obj):
    try:
        parsed = ast.literal_eval(obj)
        return [i['name'] for i in parsed]
    except:
        return []

def safe_convert_cast(obj):
    try:
        parsed = ast.literal_eval(obj)
        return [i['name'] for i in parsed[:3]]  # take top 3 cast
    except:
        return []

def safe_fetch_director(obj):
    try:
        parsed = ast.literal_eval(obj)
        for i in parsed:
            if i['job'] == 'Director':
                return [i['name']]  # keep as list
        return []
    except:
        return []

# Apply all conversions
movies['genres'] = movies['genres'].apply(safe_convert)
movies['keywords'] = movies['keywords'].apply(safe_convert)
movies['cast'] = movies['cast'].apply(safe_convert_cast)
movies['crew'] = movies['crew'].apply(safe_fetch_director)
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else [])

# Now create tags properly by concatenating lists
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Test output
movies[['title', 'tags']].head()
```

### Step 4: Cleaning and Normalizing the Tags.

```python
# Step 4: Clean and normalize the 'tags' column

# Convert list of words into a single string and lowercase everything
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))  # join the list
movies['tags'] = movies['tags'].apply(lambda x: x.lower())    # convert to lowercase
```

```python
movies[['title', 'tags']].head()
```

### Step 5: Text Vectorization using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize vectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# Transform tags column into vectors
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

### Step 6: Compute Similarity Using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# Create similarity matrix
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

###  Step 7: Build the Recommendation Function ðŸŽ¬

```python
def recommend(movie):
    # Get index of the movie from title
    movie_index = movies[movies['title'] == movie].index[0]
    
    # Get list of (index, similarity_score) for all movies
    distances = list(enumerate(similarity[movie_index]))
    
    # Sort based on similarity score in descending order
    movies_list = sorted(distances, reverse=True, key=lambda x: x[1])[1:6]
    
    # Print recommended movie titles
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
##checking the recommendation
```

```python
recommend('Avatar')
```
